```python
#visualization  aakhree pasta! plotting with respect to descriptive stats
#box and whisker
# import pandas, numpy and datetime
import numpy as np
import pandas as pd
# needed for representing dates and times
import datetime 
from datetime import datetime
# Set some pandas options for controlling output
pd.set_option('display.notebook_repr_html', False)
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows', 10)
# matplotlib and inline graphics
import matplotlib.pyplot as plt
%matplotlib inline
#generate the series
seedval=12345
np.random.seed(seedval)
dfb = pd.DataFrame(np.random.randn(10,5))
# generate the plot
dfb.boxplot(return_type='axes');
```

```python
# a bar plot# create a small series of 10 random values centered at 0.0
np.random.seed(seedval)
s = pd.Series(np.random.rand(10) - 0.5)
# plot the bar chart
s.plot(kind='bar');
```

```python
#multi sereis plot chart
# draw a multiple series bar chart
# generate 4 columns of 10 random values
np.random.seed(seedval)
df2 = pd.DataFrame(np.random.rand(10, 4),columns=['a', 'b', 'c', 'd'])
# draw the multi-series bar chart
#df2.plot(kind='bar');
# horizontal stacked bar chart
#df2.plot(kind='bar', stacked=True);
# horizontal stacked bar chart
df2.plot(kind='barh', stacked=True);
```

```python
#hostogram with  bins
   # create a histogram
np.random.seed(seedval)
# 1000 random numbers
dfh = pd.DataFrame(np.random.randn(1000))
   # draw the histogram
dfh.hist(bins=100);
```

```python
#Scatter plot
# generate a scatter plot of two series of normally distributed random values
# we would expect this to cluster around 0,0
np.random.seed(111111)
sp_df = pd.DataFrame(np.random.randn(10000, 2),columns=['a', 'b'])
sp_df.plot(kind='scatter', x='a', y='b')
```

```python
#kernel Denstity estimate plotting important in Pandas
   # create a kde density plot
   # generate a series of 1000 random numbers
np.random.seed(seedval)
s = pd.Series(np.random.randn(1000))
# generate the plot
s.hist(normed=True) # shows the bars
s.plot(kind='kde');
```
